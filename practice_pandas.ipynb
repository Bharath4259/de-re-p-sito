# Which year has the least number of stops?

```python
# To convert date and time columns to single column
# and make it's data type to datetime
dset['stop_datetime']  = pd.to_datetime(dset.stop_date.str.cat(dset.stop_time, sep = ' '))
```

```python
dset.dtypes
```

# How does Drug activity change by time of day?

#                or

# In which time drugs checking is more?

#####  use >>  ( drugs_related_stop,     stop_date )

```python
dset.stop_datetime[dset['drugs_related_stop'] == True].value_counts()
```

```python
dset.drugs_related_stop.mean()
```

```python
grp = dset.groupby(dset.stop_datetime.dt.hour).drugs_related_stop.mean() * 100
ax = grp.plot()
ax.set_xlabel('hours')
ax.set_ylabel('No of stops in %')
```

```python
grp = dset.groupby(dset.stop_datetime.dt.time).drugs_related_stop.mean() * 100
ax = grp.plot()
ax.set_xlabel('time')
ax.set_ylabel('No of times check conducted in %')
```

```python
grp = dset.groupby(dset.stop_datetime.dt.time).drugs_related_stop.count()
ax = grp.plot()
ax.set_xlabel('time')
ax.set_ylabel('No of times check conducted ')
```

# Do most stops occur at night?

```python
daytime_stops = dset[(dset.stop_datetime.dt.hour > 20) | (dset.stop_datetime.dt.hour < 4)].shape
nighttime_stops = dset[(dset.stop_datetime.dt.hour < 21) | (dset.stop_datetime.dt.hour > 3)].shape

print("Is no. of stops during night high: ", nighttime_stops > daytime_stops )
```
